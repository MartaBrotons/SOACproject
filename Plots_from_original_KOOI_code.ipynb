```python
def plots_sizes_ranges(data):
    ns_per_hour = np.timedelta64(1, 'h')
    real_time = data['time']/ns_per_hour # convert time to hours
    time_since_release = (real_time.values.transpose() - real_time.values[:,0]) # substract the initial time from each timeseries
    days_since_release = time_since_release/24
    # depth
    plt.figure(figsize = (15,5))
    plt.plot(days_since_release, -data['z'].T)
    plt.xlabel('Time (days)')
    plt.ylabel('Depth (m)')
    plt.grid()
    plt.show()
    # temperature
    plt.figure(figsize = (15,5))
    plt.plot(days_since_release, data['temp'].T)
    plt.ylim(22, 26)
    plt.grid()
    plt.xlabel('Time (days)')
    plt.ylabel('Temperature (C)')
    plt.show()
    # ambient algae concentration
    plt.figure(figsize = (15,5))
    plt.plot(days_since_release, data['aa'].T)
    plt.grid()
    plt.xlabel('Time (days)')
    plt.ylabel('Ambient algae concentration (no/m3)')
    plt.show()
    # number of attached algae
    plt.figure(figsize = (15,5))
    plt.plot(days_since_release, data['a'].T)
    plt.grid()
    plt.xlabel('Time (days)')
    plt.ylabel('Number attached algae (no/m2)')
    plt.show()
    # particle velocity
    plt.figure(figsize = (15,5))
    plt.plot(days_since_release, data['vs'].T)
    plt.grid()
    plt.xlabel('Time (days)')
    plt.ylabel('Particle velocity (m/s)')
    plt.show()
    
    
```
